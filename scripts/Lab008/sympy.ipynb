```python
t = sp.Symbol('t')
r = sp.Symbol('r')
K = sp.Symbol('K')
N0 = sp.Symbol('N_0')

N = sp.Function('N')

t, r, K, N0, N(t)
```

```python
N(t).diff(t)
```

```python
r * N(t) * (1 - N(t) / K)
```

```python
eq = sp.Eq(N(t).diff(t), r * N(t) * (1 - N(t) / K))
eq
```

```python
sp.dsolve(eq, N(t))
```

```python
sp.dsolve(eq, N(t), ics={N(0): N0})
```

```python
sol = sp.dsolve(eq, N(t), ics={N(0): N0}).simplify()
sol
```

```python
sol.lhs, sol.rhs
```

```python
sp.plot(sol.rhs.subs({r: 0.2, K: 100, N0: 10}), (t, 0, 50), title='Logistic Growth', ylabel='N(t)', xlabel='Time t')
```

```python
solf = sp.lambdify(t, sol.rhs.subs({r: 0.2, K: 100, N0: 10}))
```

```python
solf(3), type(solf(3))
```

```python
ts = np.linspace(0, 50, 1000)
plt.plot(ts, solf(ts))
plt.title('Logistic Growth')
plt.xlabel('Time t')
plt.ylabel('N(t)')
plt.grid(ls='--')
```
